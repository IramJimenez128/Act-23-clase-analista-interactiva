```python
class Analista:
    # 1. Atributos (Lo que el analista 'tiene')
    def __init__(self, nombre, especialidad):
        self.nombre = nombre
        self.especialidad = especialidad
        self.energia = 100

    # 2. Funciones/Métodos (Lo que el analista 'hace')
    def presentarse(self):
        print(f"\n😎 ¡Hola! Soy {self.nombre}, experto en {self.especialidad}.")
        print(f"Mi nivel de energía actual es: {self.energia}%")

    def analizar_tendencia(self, tema):
        print(f"\n🔍 {self.nombre} está analizando por qué '{tema}' es viral...")
        self.energia -= 20
        print(f"✅ Análisis completo. Energía restante: {self.energia}%")

    def recargar_cafe(self):
        print(f"\n☕ {self.nombre} se tomó un café cargado... ¡Pum! Energía al 100.")
        self.energia = 100

# --- LÓGICA INTERACTIVA ---

print("--- BIENVENIDO AL SIMULADOR DE ANALISTAS ---")
nom1 = input("Nombre del primer analista: ")
esp1 = input("¿En qué es experto?: ")

# Objeto 1
analista1 = Analista(nom1, esp1)

# Objeto 2 (Uno predeterminado para comparar)
analista2 = Analista("Skynet", "Predecir el futuro")

# Interacción con el Objeto 1
analista1.presentarse()
tema_interes = input(f"\n¿Qué tema quieres que {analista1.nombre} analice?: ")
analista1.analizar_tendencia(tema_interes)

# Interacción con el Objeto 2
print("\n--- MIENTRAS TANTO EL OTRO ANALISTA ---")
analista2.presentarse()
analista2.recargar_cafe()

print("\n🚀 ¡Simulación terminada!")
print("Programa realizado por Iram Jimenez NL = 28")
```
